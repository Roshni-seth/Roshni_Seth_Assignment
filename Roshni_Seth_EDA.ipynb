Univariate Analysis

1. Customer Dataset

```python
# Region Analysis

import seaborn as sns
import matplotlib.pyplot as plt

# Region distribution
sns.countplot(x='Region', data=Cust)
plt.title('Distribution of Customers by Region')
plt.xticks(rotation=45)
plt.show()
```

```python
# Signup date Analysis
Cust['SignupYear'] = Cust['SignupDate'].dt.year
Cust['SignupMonth'] = Cust['SignupDate'].dt.month

signup_counts = Cust['SignupYear'].value_counts().sort_index()
signup_counts.plot(kind='bar')
plt.title('Number of Signups by Year')
plt.xlabel('Year')
plt.ylabel('Number of Signups')
plt.show()

signup_counts = Cust['SignupMonth'].value_counts().sort_index()
signup_counts.plot(kind='bar')
plt.title('Number of Signups by Months')
plt.xlabel('Year')
plt.ylabel('Number of Signups')
plt.show()
```

2. Products Dataset

```python
# On Category Column
sns.countplot(x='Category', data=Prod)
plt.title('Product Distribution by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# On price column
sns.histplot(Prod['Price'], kde=True)
plt.title('Distribution of Product Prices')
plt.show()
```

```python
# Category wise price distribution
sns.boxplot(x='Category', y='Price', data=Prod, palette='Set2',hue='Category',legend=False)
plt.title('Price Distribution by Product Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 expensive products
top_expensive_products = Prod.nlargest(10, 'Price')
sns.barplot(x='Price', y='ProductName', data=top_expensive_products, palette='coolwarm',hue='ProductName', legend=False)
plt.title('Top 10 Most Expensive Products')
plt.show()
```

3. Transactions Dataset

```python
# Transaction value distribution
sns.histplot(Trans['TotalValue'], kde=True)
plt.title('Distribution of Transaction Values')
plt.show()
```

```python
# Revenue trends overtime
Trans['TransactionDate'] = pd.to_datetime(Trans['TransactionDate'])
revenue_trends = Trans.groupby(Trans['TransactionDate'].dt.date)['TotalValue'].sum()

revenue_trends.plot(title='Revenue Trends Over Time', figsize=(10, 6))
plt.xlabel('Transaction Date')
plt.ylabel('Total Revenue')
plt.grid()
plt.show()
```

```python
# Top 10 customers by Revenue
top_customers = Trans.groupby('CustomerID')['TotalValue'].sum().nlargest(10).reset_index()

sns.barplot(x='TotalValue', y='CustomerID', data=top_customers, palette='plasma',hue='CustomerID',legend=False)
plt.title('Top 10 Customers by Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Customer ID')
plt.show()
```

```python
print((((Trans.groupby('CustomerID')['TotalValue'].sum().nlargest(10).reset_index()).sum())))
```

```python
# Category-Wise Revenue distribution
transactions_products = Trans.merge(Prod, on='ProductID', how='inner')
category_revenue = transactions_products.groupby('Category')['TotalValue'].sum().reset_index()

sns.barplot(x='Category', y='TotalValue', data=category_revenue, palette='pastel', hue='Category',legend=False)
plt.title('Revenue Contribution by Product Category')
plt.xticks(rotation=45)
plt.show()
```

Bivariate Analysis

```python
# Transaction by Customer Region
transactions_customers = Trans.merge(Cust, on='CustomerID')
region_sales = transactions_customers.groupby('Region')['TotalValue'].sum()

region_sales.plot(kind='bar', title='Total Sales by Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Quantity vs Price
sns.scatterplot(x='Price_y', y='Quantity', data=transactions_products)
plt.title('Quantity vs Price')
plt.show()
```

MultiVariate Analysis

```python
# Revenue by Region and Categories
transactions_full = transactions_products.merge(Cust, on='CustomerID', how='inner')
region_category_revenue = transactions_full.groupby(['Region', 'Category'])['TotalValue'].sum().unstack()

region_category_revenue.plot(kind='bar', stacked=True, figsize=(12, 8), cmap='tab20')
plt.title('Revenue by Region and Product Category')
plt.ylabel('Total Revenue')
plt.xlabel('Region')
plt.legend(title='Category')
plt.show()
```

```python
sns.scatterplot(x='Quantity', y='TotalValue', data=Trans, alpha=0.7, color='purple')
plt.title('Transaction Quantity vs Total Value')
plt.xlabel('Quantity')
plt.ylabel('Total Value')
plt.show()
```

```python
# Top 10 products by quantity sold
top_products = transactions_products.groupby('ProductName')['Quantity'].sum().nlargest(10).reset_index()

sns.barplot(x='Quantity', y='ProductName', data=top_products, palette='coolwarm')
plt.title('Top 10 Products by Quantity Sold')
plt.xlabel('Quantity Sold')
plt.ylabel('Product Name')
plt.show()
```

```python
# Average price vs Quantity for Categories
avg_price_quantity = transactions_products.groupby('Category').agg({'Price_y': 'mean', 'Quantity': 'sum'}).reset_index()

sns.scatterplot(x='Price_y', y='Quantity', hue='Category', data=avg_price_quantity, palette='Dark2', s=100)
plt.title('Average Price vs Total Quantity Sold by Category')
plt.xlabel('Average Price')
plt.ylabel('Total Quantity Sold')
plt.legend(title='Category')
plt.show()
```
